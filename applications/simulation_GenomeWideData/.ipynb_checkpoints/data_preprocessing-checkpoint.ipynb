```r
suppressPackageStartupMessages({
    library(tidyverse)
    library(anndata)
    library(Seurat)   
})
```

# load mixed data, gene x cell format, with an extra celltype column

```r
raw_data <- readRDS('rds/NT.essential.raw.k562_rpe1_combine_type.rds')
```

```r
raw_data%>%.[1:4,c(1:3,ncol(raw_data))]
```

```r
mat.combine <- raw_data%>%select(!celltype)

mat.label <- raw_data%>%select(celltype)
```

# create seurat object, visulize data and apply quality control

```r
seurat <- CreateSeuratObject(mat.combine%>%t%>%as.data.frame,meta.data=mat.label,min.cells = 100,min.features = 500)
```

```r
seurat
```

```r
seurat <- NormalizeData(seurat, normalization.method = "LogNormalize", scale.factor = 10000)
```

```r
# Visualize QC metrics as a violin plot
VlnPlot(seurat, features = c("nFeature_RNA", "nCount_RNA"), ncol = 2)
```

```r
# FeatureScatter is typically used to visualize feature-feature relationships, but can be used
# for anything calculated by the object, i.e. columns in object metadata, PC scores etc.

plot1 <- FeatureScatter(seurat, feature1 = "nCount_RNA", feature2 = "nFeature_RNA")
plot1
```

```r
seurat <- subset(seurat, 
                 subset = nFeature_RNA > 2000 & 
                         nFeature_RNA < 5000 &
                         nCount_RNA <40000)
```

```r
seurat
```

## find varible genes

```r
seurat <- FindVariableFeatures(seurat, selection.method = "vst", nfeatures = 2000)
```

```r
# Identify the 10 most highly variable genes
top10 <- head(VariableFeatures(seurat), 10)

# plot variable features with and without labels
plot1 <- VariableFeaturePlot(seurat)
plot2 <- LabelPoints(plot = plot1, points = top10, repel = TRUE)
```

```r
plot2
```

## visulize data distribution

```r
all.genes <- rownames(seurat)
seurat <- ScaleData(seurat, features = all.genes)

seurat <- RunPCA(seurat, features = VariableFeatures(object = seurat))
```

```r
DimPlot(seurat, reduction = "pca",group.by = 'celltype')
```

## save essential processed data to rds

```r
data.norm <- as.matrix(seurat@assays$RNA@data)
features.select <- VariableFeatures(object = seurat)

# calculate gene variance within each celltype
k562.gene.variance <- data.norm[,seurat@meta.data%>%filter(celltype=='k562')%>%rownames]%>%apply(1,var)
rpe1.gene.variance <- data.norm[,seurat@meta.data%>%filter(celltype=='rpe1')%>%rownames]%>%apply(1,var)
```
